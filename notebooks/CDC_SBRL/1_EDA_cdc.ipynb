## Determine the availability of each assay

```python
from functools import reduce
def get_available_data(cdc_data, cols_to_use, value_options):
    valid_cnt = []
    valid_df = []
    for col in cols_to_use:
        valid_temp = cdc_data[cdc_data[col].isin(value_options)][col].reset_index()
        valid_df.append(valid_temp)
        valid_cnt.append(len(valid_temp)/len(cdc_data))
    joined_valid_df = reduce(lambda left, right: pd.merge(left, right, on = 'index', how = 'outer'), valid_df)
    valid_cnt_df = pd.DataFrame(valid_cnt, index = cols_to_use)
    plot_available_data(valid_cnt_df)
    return joined_valid_df
```

```python
def plot_available_data(valid_cnt_df):
    plt.figure(figsize = (1, len(valid_cnt_df)/1.5))
    sns.heatmap(valid_cnt_df, cmap = 'Reds', center = 0.5, vmin = 0, vmax = 1, annot = True)
    plt.yticks(rotation = 0)
    plt.xticks([])
    plt.title('Percentage of available data')
```

## Plot heatmap to see if friends and foes have distinct assay results

```python
def plot_assay_group_by_foe(assay_data, cols_to_use):
    assay_value_cnt_dfs = []
    for col in cols_to_use:
        value_cnt = assay_data.groupby(['Foe'])[col].value_counts(normalize = True)
        value_cnt.index.names = ['Foe', 'type'] # rename index so could reset_index
        value_cnt = value_cnt.reset_index()
        value_cnt['test'] = col # add a new column for the test type
        value_cnt.rename(columns = {col:'values'}, inplace=True) # rename the original column to values
        assay_value_cnt_dfs.append(value_cnt)
    return pd.concat(assay_value_cnt_dfs).reset_index(drop = True)
```

```python
def draw_heatmap(*args, **kwargs):
    sns.set(font_scale = 1.25)
    sns.set_style('white')
    data = kwargs.pop('data')
    d = data.pivot(index=args[1], columns=args[0], values=args[2])
    sns.heatmap(d, **kwargs)
```

```python
sugar_value_cnt = plot_assay_group_by_foe(infrequent_only_w_labels, sugar_cols)
g = sns.FacetGrid(sugar_value_cnt, col='test', col_wrap=3)
g.map_dataframe(draw_heatmap, 'Foe', 'type', 'values', cmap = 'Reds', vmin = 0, vmax = 1, center = 0.5)
```

```python
ferment_value_cnt = plot_assay_group_by_foe(infrequent_only_w_labels, ferment_cols)
g = sns.FacetGrid(ferment_value_cnt, col='test', col_wrap=3)
g.map_dataframe(draw_heatmap, 'Foe', 'type', 'values', cmap = 'Reds', vmin = 0, vmax = 1, center = 0.5)
```

### Convert the data from results to ratio of each kind of results, like '-' or '+' in 'MaC/SS = SS' to ratios in 'MaC/SS = SS_-' and 'MaC/SS = SS_+'

```python
# make new columns with all the possible assay results
new_cols = []
for col in clean_valid_genus_species.columns[:-2]:
    for category in clean_valid_genus_species[col].unique():
        new_cols.append(f'{col}_{category}')
new_cols.append(f'instance_sum')
```

```python
# new df with the columns as all the assay results
one_hot_df = pd.DataFrame(columns = new_cols, index = clean_valid_genus_species['final_id_clean'].unique())
```

```python
# iterate the strains and assay results to count
for col in clean_valid_genus_species.columns[:-2]:
    for category in clean_valid_genus_species[col].unique():
        new_col = f'{col}_{category}'
        cnt_each_strain = []
        for strain in clean_valid_genus_species['final_id_clean'].unique():
            temp = clean_valid_genus_species[(clean_valid_genus_species['final_id_clean'] == strain) & (clean_valid_genus_species[col] == category)] # get rows with specific assay results
            cnt_each_strain.append(len(temp)) # add the result count
        one_hot_df[new_col] = cnt_each_strain
```
